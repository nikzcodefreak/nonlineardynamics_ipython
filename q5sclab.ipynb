```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X,Y=np.meshgrid(np.arange(-2*np.pi,2*np.pi,.5),np.arange(-2*np.pi,2*np.pi,.5))
```

```python
#Question5a:
dx=Y
dy=-2*X-3*Y
```

```python
#nullclines
xn=np.arange(-2*np.pi,2*np.pi,.5)
yn1=np.zeros(len(xn))

yn2=-0.667*xn
plt.figure()
plt.plot(xn,yn1,xn,yn2)
Q=plt.quiver(X,Y,dx,dy)
plt.show()
```
