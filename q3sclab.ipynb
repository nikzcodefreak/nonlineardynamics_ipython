```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X,Y=np.meshgrid(np.arange(-2*np.pi,2*np.pi,.5),np.arange(-2*np.pi,2*np.pi,.5))
```

```python
dx=-2*X*Y
dy=(2*X*Y-5*Y)
```

```python
plt.figure()
xn=np.arange(-2*np.pi,2*np.pi,.5)
yn=np.arange(-2*np.pi,2*np.pi,.5)
yn1=np.zeros(len(xn))
xn1=np.zeros(len(xn))
xn2=0.4*np.ones(len(yn))
plt.plot(xn,yn1,xn1,yn,xn2,yn)
Q=plt.quiver(X,Y,dx,dy)
strm=plt.streamplot(X,Y,dx,dy)
plt.show()
```
