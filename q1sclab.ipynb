```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X,Y=np.meshgrid(np.arange(-2*np.pi,2*np.pi,.5),np.arange(-2*np.pi,2*np.pi,.5))
```

```python
#Question1:R=2,L=1,C=1
#lambda1=lambda2=-1
dx=Y
dy=-(2*Y+X)
```

```python
plt.figure()
L=np.sqrt((dx**2)+(dy**2))
Q=plt.quiver(X,Y,dx/L,dy/L)
strm=plt.streamplot(X,Y,dx,dy)
plt.title("star Stable")
plt.show()
```

```python
dx1=Y
dy1=-(5*Y+0.5*X)
L=np.sqrt((dx1**2)+(dy1**2))
plt.figure()
Q1=plt.quiver(X,Y,dx1/L,dy1/L)
strm1=plt.streamplot(X,Y,dx1,dy1)

plt.title("Stable")
plt.show()
```

```python
dx2=Y
dy2=-(Y+X)
plt.figure()
Q2=plt.quiver(X,Y,dx2,dy2)
strm=plt.streamplot(X,Y,dx2,dy2)

plt.title("Spiral Stable")
plt.show()
```
